```python
from prophet import Prophet
```

```python
data = data.rename(columns={ data.columns[0]: 'ds', data.columns[1]: 'y' })
```

```python
# Ensure correct data types
data['ds'] = pd.to_datetime(data['ds'])              # convert to datetime
data['y'] = pd.to_numeric(data['y'], errors='coerce') # convert to numeric
```

```python
data = data.dropna(subset=['ds', 'y'])
```

```python
model = Prophet()
model.fit(data[['ds', 'y']])
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat','yhat_lower', 'yhat_upper']].tail()
```

```python
figi = model.plot(forecast)
```

```python
figi2 = model.plot_components(forecast)
```
